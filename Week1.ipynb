```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow #for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
#membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://awsimages.detik.net.id/community/media/visual/2023/09/02/kucing-ragdoll_169.jpeg?w=600&q=90"]

#baca dan tampilkan imaga
#loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("Resolusi Image =", tinggi, "x", lebar, "\nTinggi =", tinggi, "\nLebar =", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal di tengah image
for y in range (lebar):
  image_3[int((tinggi)/2), y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

3. Kode ini menggunakan fungsi cv.resize dari OpenCV. Dalam fungsi ini, ukuran gambar tidak ditentukan secara langsung melalui parameter dimensi (lebar dan tinggi), melainkan dengan menggunakan faktor skala. Nilai (0, 0) di sini berarti ukuran akhir gambar akan dihitung berdasarkan faktor skala yang diberikan.

    * fx=0.5: Menentukan faktor skala untuk sumbu horizontal (x-axis), yang berarti gambar akan diperkecil menjadi 50% dari lebar aslinya.
    * fy=0.5: Menentukan faktor skala untuk sumbu vertikal (y-axis), yang berarti gambar akan diperkecil menjadi 50% dari tinggi aslinya.

  Dengan demikian, hasil dari kode ini adalah gambar yang ukurannya menjadi setengah dari ukuran aslinya, baik dalam arah horizontal maupun vertikal.
4. Potongan kode tersebut menggunakan format RGB (Red, Green, Blue) untuk menentukan warna, di mana [255, 255, 255] adalah kode untuk warna putih. Fungsi dari potongan kode tersebut adalah untuk mengatur warna piksel pada posisi tertentu, yakni membentuk garis horizontal putih di tengah gambar.
5.  Piksel adalah unit terkecil dari gambar digital yang menyimpan informasi warna. Resolusi gambar merujuk pada jumlah piksel dalam dimensi horizontal dan vertikal, seperti 1920x1080. Gambar dengan resolusi tinggi memiliki lebih banyak piksel, sehingga dapat menampilkan lebih banyak detail dan kejelasan. Sebaliknya, gambar dengan resolusi rendah memiliki lebih sedikit piksel, yang sering kali menghasilkan kualitas gambar yang kurang tajam. Resolusi yang lebih tinggi berarti gambar lebih detail dan jelas, sementara resolusi yang lebih rendah berarti gambar lebih buram dan kurang detail.
